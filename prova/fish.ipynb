### Questão 1 (1,0 pt)

Descreva os conceitos de *precision* e *recall* com suas palavras. Construa um exemplo de problema de classificação - que não seja nem médico, nem jurídico, para não reutilizar os exemplos de sala de aula - e discuta neste exemplo as consequências de

- um alto *precision* e baixo *recall*
- um alto *recall* e um baixo *precision*.

**Resposta:**

**Conceito de precision e recall**

Precision é uma métrica que avalia a proporção de exemplos positivos que foram corretamente classificados como positivos, em relação ao número total de exemplos classificados como positivos pelo modelo. Ela mede a capacidade do modelo de distinguir entre casos positivos e negativos.

Recall é um parâmetro utilizado para avaliar o quão precisamente um modelo de classificação de machine learning detectou amostras positivas em relação ao número total de amostras que eram verdadeiramente positivas. Recall avalia a capacidade do modelo em identificar todos os casos de sucesso.

**Consequências de um alto precision e baixo recall**

O modelo tende a fazer menos previsões positivas, mas quando faz, geralmente são precisas devido à sua alta precisão e baixo recall. Essa combinação pode sugerir que o modelo está fazendo poucas previsões e pode estar negligenciando inúmeras situações bem-sucedidas.

**Exemplo de modelo com alto precision e baixo recall:**

O checkpoint de segurança do aeroporto possui um detector de metal. Devido à excelente precisão desse detector, ele frequentemente emite um som quando encontra metal em uma pessoa. No entanto, o seu baixo recall faz com que não todos os objetos de metal nas pessoas que passam pelo checkpoint sejam detectados. Devido a isso, mesmo que seja eficaz em encontrar objetos metálicos, ele pode perder alguns, o que pode ser um problema se alguém estiver carregando algo perigoso como uma arma ou explosivo.

**Consequências de alto recall e baixo precision**

Ter um alto recall e baixa precisão significa que o modelo está reconhecendo muitos casos positivos, mas muitas dessas identificações estão incorretas. Essa combinação pode indicar que o modelo está fazendo previsões excessivamente generosas e pode estar incluindo um grande número de casos errôneos.

**Exemplo de alto recall e baixo precision**

O ponto de segurança do aeroporto utiliza um modelo de aprendizagem de máquina para identificar mercadorias proibidas nas bagagens. Com base no tamanho, forma e densidade dos itens, o modelo foi treinado para identificá-los. O modelo consegue identificar a maioria das mercadorias proibidas nas malas que passam pelo ponto de segurança devido a sua alta taxa de recall. Entretanto, devido à sua baixa precisão, o modelo frequentemente classifica erroneamente mercadorias lícitas como ilegais, levando a um número significativo de alertas falsos.


### Questão 2: (1,0 pt)

Em um problema de *machine learning* temos uma *feature* discreta com os seguintes valores possíveis: "Aprendiz", "Junior", "Senior", "Especialista". Estas categorias estão em ordem: um "Aprendiz" evolui para um "Junior", etc. Como esta *feature* deve ser representada em um projeto de regressão linear? Explique vantagens e desvantagens da representação que você escolheu. Dica: não tem só uma resposta certa aqui, mas não preciso de todas as respostas certas possíveis não, apenas de uma que você escolher.

**Resposta:**

No aprendizado de máquina, às vezes usamos variáveis categóricas ordenadas (como "Aprendiz", "Junior", "Senior" e "Especialista"), em que um valor é maior que outro (Aprendiz Junior Senior Especialista).
Para usar essa variável em um modelo de regressão linear, ela deve ser representada como valores numéricos. Uma escolha é usar uma escala numérica (1 para Aprendiz, 2 para Junior, etc.). Outra escolha é usar variáveis binárias, chamadas de "variáveis dummy", em que estabelecemos uma variável binária distinta para cada categoria (como uma variável chamada "Aprendiz" que é 1 se a variável original era Aprendiz e 0 caso contrário).
Usar uma escala numérica é mais fácil e compacto do que usar variáveis dummy, o que é uma vantagem. No entanto, a suposição de que a distinção entre as categorias é a mesma (por exemplo, que a distinção entre Aprendiz e Junior é a mesma que a distinção entre Junior e Senior) pode não ser precisa na prática.

### Questão 4: (1,0 pt)

Suponha que você está explicando *machine learning* *para seus colegas do primeiro ano, fazendo ciência dos dados*. Explique como funciona o *gradient descent* na busca de melhores parâmetros para um modelo linear.

**Resposta:**
A descida de gradiente é um algoritmo simples: começamos com um conjunto de parâmetros aleatórios e iterativamente os atualizamos na direção do gradiente decrescente da função de perda (loss function). A função de perda mede o quão bem ou mal nosso modelo faz a previsão com os parâmetros atuais.
Na prática, isso significa que, a cada iteração, calculamos o gradiente da função de perda em relação aos parâmetros e, em seguida, atualizamos os parâmetros multiplicando-os por uma taxa de aprendizado (learning rate) para movê-los na direção oposta ao gradiente. Essa taxa de aprendizado regula o tamanho da atualização dos parâmetros em cada iteração, para evitar atualizações muito grandes ou muito pequenas.
Assim, a descida de gradiente nos ajuda a determinar iterativamente os parâmetros ideais para nosso modelo linear, ajustando-os até que a função de perda seja minimizada e nosso modelo produza a melhor previsão possível

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train_valid, X_test_valid, y_train_valid, y_test_valid = train_test_split(X_train, y_train, test_size=0.33)
```

```python
import math
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train_valid, y_train_valid)

y_pred = model.predict(X_test_valid)
RMSE = math.sqrt(mean_squared_error(y_test_valid, y_pred))
RMSE
```

```python
model = LinearRegression()

model.fit(X_train_valid_transf, y_train_valid)
```

```python
X_test_valid_transf = transformer.transform(X_test_valid)
y_pred = model.predict(X_test_valid_transf)
```
